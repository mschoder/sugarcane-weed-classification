## Edge Detection

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import math
```

```python
def edge_detection(filepath):
    print(filepath)
    im = cv2.imread(filepath)
    im_rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)

    gray = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE)

    # blurring to get rid of the noise; invert image
    # gray_blur = 255 - cv2.medianBlur(gray, 3)
    gray_blur = 255 - cv2.GaussianBlur(gray, (5, 5), 1.4) 
#     gray_blur = cv2.bilateralFilter(gray, 7, 50, 50)

    # Canny edge detection
    edges = cv2.Canny(gray_blur, 50, 150, apertureSize = 3)

    # Initialize output
    out = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)

    # Detect and draw lines
    lines = cv2.HoughLinesP(edges, 1, np.pi/180, 10, minLineLength=65, maxLineGap=10)
    for line in lines:
        for x1, y1, x2, y2 in line:
            cv2.line(out, (x1, y1), (x2, y2), (255, 0, 0), 2)

    # gray_blur_rgb = cv2.cvtColor(gray, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(16,10))
    plt.subplot(141)
    plt.title("Original Image")
    plt.imshow(im_rgb)

    plt.subplot(142)
    plt.title("Grayscale Original")
    plt.imshow(gray, cmap = 'gray')

    plt.subplot(143)
    plt.title("Canny Edges")
    plt.imshow(edges)

    plt.subplot(144)
    plt.title("Hough Lines")
    plt.imshow(out, cmap='gray')
```
